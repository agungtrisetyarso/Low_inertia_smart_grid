```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt

# ========================== PARAMETERS ==========================
omega1 = 3.83          # rad/s
omega2 = 3.90          # rad/s
J_intra = 2.8          # intra-pair coupling
J_inter = 1.5          # inter-pair (automation task-substitution) coupling
R = 0.15               # reskilling parameter
T = 20.0
n_points = 400
t = np.linspace(0, T, n_points)

np.random.seed(42)

# ========================== 4-QUBIT OPERATORS ==========================
sz = [qt.tensor(*[qt.sigmaz() if i==k else qt.qeye(2) for k in range(4)]) for i in range(4)]
sx = [qt.tensor(*[qt.sigmax() if i==k else qt.qeye(2) for k in range(4)]) for i in range(4)]
sy = [qt.tensor(*[qt.sigmay() if i==k else qt.qeye(2) for k in range(4)]) for i in range(4)]

H = (omega1 * (sz[0] + sz[1]) + omega2 * (sz[2] + sz[3]) +
     J_intra * (sx[0]*sx[1] + sy[0]*sy[1] + sz[0]*sz[1]) +
     J_intra * (sx[2]*sx[3] + sy[2]*sy[3] + sz[2]*sz[3]) +
     J_inter * (sx[0]*sx[2] + sy[0]*sy[2] + sz[0]*sz[2]) +
     R * (sz[0] + sz[1] + sz[2] + sz[3]))

# ========================== SEPARABLE INITIAL STATE ==========================
psi0 = qt.tensor(
    (qt.basis(2,0) + qt.basis(2,1)).unit(), qt.basis(2,0),
    (qt.basis(2,0) + qt.basis(2,1)).unit(), qt.basis(2,0)
)

# ========================== TIME EVOLUTION ==========================
result = qt.mesolve(H, psi0, t)

# ========================== ENTANGLEMENT & STABILITY INDEX ==========================
conc = []
entropy = []
I_stab = []

for state in result.states:
    rho12 = qt.ptrace(state, [0,1])                    # pair 1-2
    rho1  = qt.ptrace(state, [0])                      # single qubit
    conc.append(qt.concurrence(rho12))
    S_vN = qt.entropy_vn(rho1)
    entropy.append(S_vN)
    I_stab.append(1 - S_vN / np.log(2))                # Grid Information Stability Index

# ========================== PLOT 1: Time evolution (recommended for manuscript) ==========================
fig, ax1 = plt.subplots(figsize=(10, 6))

color1 = 'tab:blue'
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Concurrence (pair 1-2) / von Neumann entropy', color=color1)
ax1.plot(t, conc, 'b-', linewidth=2.5, label='Concurrence (pair 1-2)')
ax1.plot(t, entropy, 'r--', linewidth=2.5, label='von Neumann entropy (qubit 1)')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, alpha=0.3)

ax2 = ax1.twinx()
color2 = 'tab:green'
ax2.set_ylabel('Grid Information Stability Index $\mathcal{I}_{\\rm stab}$', color=color2)
ax2.plot(t, I_stab, 'g-', linewidth=3, label='Grid Information Stability Index')
ax2.tick_params(axis='y', labelcolor=color2)

fig.suptitle('4-Qubit Case: Dynamic Entanglement and Grid Information Stability Index\nfrom Separable Initial State', fontsize=14)
fig.legend(loc='upper right', bbox_to_anchor=(0.85, 0.85))
plt.tight_layout()
plt.savefig('fig_gisi_time.png', dpi=300, bbox_inches='tight')
plt.show()
```
